```python
import requests
import pandas as pd
import numpy as np

r = requests.get("https://raw.githubusercontent.com/pesikj/progr2-python/master/data/titanic.csv")
open("titanic.csv", 'wb').write(r.content)

data = pd.read_csv("titanic.csv")

data
```

```python

titanic_pivot = pd.pivot_table(data, index="Pclass", columns="Sex", values="Survived", aggfunc= (['sum',len, 'mean']), margins=True).reset_index().rename(columns={'sum': 'total survivers', 'len': 'total passangers'})

pd.options.display.float_format = '{:,.1%}'.format
titanic_pivot
```
